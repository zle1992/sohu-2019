```python
def read_json(path,flag='train'):
    print(path)
    out_path = path+'.pick'

    load_dicts =[]


    entity_list = []
    emotion_list= []
    with open(path,'r') as f:
       
        for i ,jsondata in enumerate(f.readlines()):
            load_dict = json.loads(jsondata)
            if flag=='test':
                load_dict['coreEntityEmotions'] = [{'entity':'null','emotion':'null'}]

            entity = [i['entity'] for i in load_dict['coreEntityEmotions']]
            emotion = [i['emotion'] for i in load_dict['coreEntityEmotions']]

            entity_list.append(entity)
            emotion_list.append(emotion)
            load_dicts.append(load_dict)

    df = pd.DataFrame(load_dicts)

    #df['texts'] = df['title'] +'-'+df['title']

    df['entity'] = df['coreEntityEmotions'].map(lambda x: ' '.join([i['entity'] for i in x]))
    df['emotion'] = df['coreEntityEmotions'].map(lambda x: ' '.join([i['emotion'] for i in x]))
    return df
```
